### libraries

```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
from sklearn import linear_model
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from sklearn import svm
from sklearn import tree
from sklearn.ensemble import RandomForestRegressor
```

### Import Data Set

```python
df = pd.read_csv("Data.csv")
print("Data")
print(df)

x= df.iloc[:,:-1].values
y=df.iloc[:,-1].values

print("Features")
print(x)

print("Dependent Variable")
print(y)
```

### Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=1)
```

```python
print("training set features")
print(len(x_train))
print("training set dependent variables")
print(len(y_train))

print("Test set features")
print(len(x_test))
print("Test set dependent variables")
print(len(y_test))
```

```python
poly = PolynomialFeatures(degree=2)
print(x_train[0:5,:])
x_poly=poly.fit_transform(x_train)
print(x_poly[0:5,:])
```

```python
def evalModel(y_test,y_pred):
  from sklearn.metrics import r2_score
  return r2_score(y_test, y_pred)
```

### Training Different Models

#### multipleRegression

```python
def multipleRegression(x_train,y_train,x_test,y_test):
  print("multipleRegression ")

  regressor=linear_model.LinearRegression()
  regressor.fit(x_train,y_train)
  y_predict=regressor.predict(x_test)
  print(np.concatenate([y_test.reshape(-1,1),y_predict.reshape(-1,1)],axis=1))

  r2Score=evalModel(y_test=y_test,y_pred=y_predict)
  print("r2Score ")

  print(r2Score)
  return ("multipleRegression",r2Score)
```

#### polynomialRegression

```python
def polynomialRegression(x_train,y_train,x_test,y_test):
  print("polynomialRegression ")

  poly = PolynomialFeatures(degree=4)
  x_train_poly=poly.fit_transform(x_train)
  regressor=linear_model.LinearRegression()
  regressor.fit(x_train_poly,y_train)

  y_predict=regressor.predict(poly.transform(x_test))
  print(np.concatenate([y_test.reshape(-1,1),y_predict.reshape(-1,1)],axis=1))

  r2Score=evalModel(y_test=y_test,y_pred=y_predict)
  print("r2Score ")

  print(r2Score)
  return ("polynomialRegression",r2Score)
```

#### supportVectorRegression

```python
def supportVectorRegression(x_train,y_train,x_test,y_test):
  print("supportVectorRegression ")

  # feature scaling for features and dependent variables
  scaler_x = preprocessing.StandardScaler().fit(x_train)
  scaler_y = preprocessing.StandardScaler().fit(y_train.reshape(-1,1))

  x_scaled=scaler_x.transform(x_train)
  y_scaled=scaler_y.transform(y_train.reshape(-1,1))

  print("x_scaled")
  print(x_scaled)
  print("y_scaled")
  print(y_scaled)


  regr = svm.SVR()
  regr.fit(x_scaled,y_scaled)

  y_predict= scaler_y.inverse_transform(regr.predict(scaler_x.transform(x_test)).reshape(-1,1))
  print(np.concatenate([y_test.reshape(-1,1),y_predict.reshape(-1,1)],axis=1))


  r2Score=evalModel(y_test=y_test,y_pred=y_predict)
  print("r2Score ")

  print(r2Score)
  return ("supportVectorRegression",r2Score)
```

#### decisionTreeRegression

```python
def decisionTreeRegression(x_train,y_train,x_test,y_test):
  print("decisionTreeRegression ")

  regressor=tree.DecisionTreeRegressor()
  regressor.fit(x_train,y_train)

  y_predict=regressor.predict(x_test)
  print(np.concatenate([y_test.reshape(-1,1),y_predict.reshape(-1,1)],axis=1))

  r2Score=evalModel(y_test=y_test,y_pred=y_predict)
  print("r2Score ")

  print(r2Score)
  return ("decisionTreeRegression",r2Score)
```

#### randomForestsRegression

```python
def randomForestsRegression(x_train,y_train,x_test,y_test):
  print("randomForestsRegression ")

  regressor= RandomForestRegressor(n_estimators=100)
  regressor.fit(x_train,y_train)

  y_predict=regressor.predict(x_test)
  print(np.concatenate([y_test.reshape(-1,1),y_predict.reshape(-1,1)],axis=1))


  r2Score=evalModel(y_test=y_test,y_pred=y_predict)
  print("r2Score ")

  print(r2Score)
  return ("randomForestsRegression",r2Score)
```

```python
regressionModels = [multipleRegression , polynomialRegression, supportVectorRegression, decisionTreeRegression , randomForestsRegression]
```

### Finding the Best Model

```python
def findBestPerformingRegressionModel():
  bestPerforming = []
  for func in regressionModels:
    print("-----------------------")
    result=func(x_train,y_train,x_test,y_test)
    modelName, score = result

    if(not bestPerforming):
      bestPerforming= [modelName,score]
    elif(score > bestPerforming[1]):
      bestPerforming= [modelName,score]

    print("-----------------------")

  return bestPerforming
```

```python
result= findBestPerformingRegressionModel()
```
