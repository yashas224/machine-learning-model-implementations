### libraries

```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
```

### Import Data Set

```python
df = pd.read_csv("Position_Salaries.csv")
print("Data")
print(df)

x= df.iloc[:,:-1].values
y=df.iloc[:,-1].values

print("Features")

print(x)

print("Dependent Variable")

print(y)
#  leaving the first column as 2nd column is encoded version of the first one
x= df.iloc[:,1:-1].values

print("FInal Features")

print(x)
```

###Training Linear regression model on the training set



```python
from sklearn.linear_model import LinearRegression
# Create model
linearRegression = LinearRegression()

# Train model
linearRegression.fit(x, y)

print(linearRegression.coef_)
print(linearRegression.intercept_)

```

### Training Polynomial Regression on dataset

```python
from sklearn.preprocessing import PolynomialFeatures
poly_regression = PolynomialFeatures(degree=4 )
x_poly=poly_regression.fit_transform(x)
print(x_poly)

polynomialRegression = LinearRegression()
polynomialRegression.fit(x_poly,y)

print(polynomialRegression.coef_)
print(polynomialRegression.intercept_)
```

### visualization Linear regression results


```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='red')
plt.plot(x, linearRegression.predict(x), color='blue')
plt.title('Linear regression')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

### visualization Polynomial regression results


```python
import matplotlib.pyplot as plt
plt.scatter(x, y, color='red')
plt.plot(x, polynomialRegression.predict(x_poly), color='blue')
plt.title('Polynomial regression')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

### Predicting salary for Linear regression

```python
linearRegression.predict([[6.5]])
```

### Predicting salary for Polynomial regression

```python
polynomialRegression.predict(poly_regression.fit_transform([[6.5]]))
```
