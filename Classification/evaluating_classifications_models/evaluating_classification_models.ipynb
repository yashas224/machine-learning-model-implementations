###imports

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Datasets

```python
df = pd.read_csv('Data.csv')
print(df)
print(df.shape)
```

```python
x=df.iloc[:,:-1].values
y=df.iloc[:,-1].values
```

### splitting data

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size= 0.25,random_state=0)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled=scaler.transform(x_test)
```

### Confusion metrix and Accuracy

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
def evaluateClassificationModel(y_test,y_predicted):
  cm = confusion_matrix(y_test, y_predicted)
  score=accuracy_score(y_test, y_predicted)
  print('Confusion Matrix')
  print(cm)
  print('Accuracy Score')
  print(score)
  return score
```

```python
def prettyPrint(modelName):
  print('--------------------------------------------------')
  print(f'{modelName}')
```

### Training  Classifier models

```python
from sklearn.ensemble import RandomForestClassifier

def randomForestClassifier():
  prettyPrint('randomForestClassifier')

  classifier = RandomForestClassifier(n_estimators=100,criterion='entropy',random_state=0)
  # train the model
  classifier.fit(x_train_scaled,y_train)
  # predict test set result
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)
```

```python
from sklearn.tree import DecisionTreeClassifier

def decissionTreeClasifier():
  prettyPrint('decissionTreeClasifier')
  classifier = DecisionTreeClassifier(criterion='entropy' , random_state=0)
  classifier.fit(x_train_scaled,y_train)
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)
```

```python
from sklearn.neighbors import KNeighborsClassifier

def kNearestNeighbourClassifier():
  prettyPrint('kNearestNeighbourClassifier')
  classifier = KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
  classifier.fit(x_train_scaled,y_train)
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)

```

```python
from sklearn.svm import SVC

def kernalSVMNonLinear():
  prettyPrint('kernalSVMNonLinear')
  classifier = SVC(kernel='rbf')
  classifier.fit(x_train_scaled,y_train)
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)
```

```python
from sklearn.linear_model import LogisticRegression

def logisticRegressionClassification():
  prettyPrint('logisticRegressionClassification')
  classifier = LogisticRegression()
  classifier.fit(x_train_scaled,y_train)
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)
```

```python
from sklearn.naive_bayes import GaussianNB

def naiveBayesClassification():
  prettyPrint('naiveBayesClassification')

  classifier = GaussianNB()
  classifier.fit(x_train_scaled,y_train)
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)
```

```python
from sklearn.svm import SVC

def SVMLinearClassifier():
  prettyPrint('SVMLinearClassifier')

  classifier = SVC(kernel='linear')
  classifier.fit(x_train_scaled,y_train)
  y_predicted=classifier.predict(x_test_scaled)
  return evaluateClassificationModel(y_test,y_predicted)
```

### Evaluating All Classification Models

```python
classificationModels = [randomForestClassifier, decissionTreeClasifier, kNearestNeighbourClassifier, kernalSVMNonLinear, logisticRegressionClassification, naiveBayesClassification, SVMLinearClassifier]
```

```python
maxAccuracy= 0.0
maxAccuracyModelName=''
for func in classificationModels:
  modelResult = func()
  if(modelResult > maxAccuracy):
    maxAccuracy= modelResult
    maxAccuracyModelName=func.__name__

print("Best Classification Model:")
print(maxAccuracy, maxAccuracyModelName)
```
