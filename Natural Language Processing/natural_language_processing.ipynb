###imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import re
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
from nltk.stem import PorterStemmer
from sklearn.model_selection import train_test_split

```

### Datasets

```python
from csv import QUOTE_NONE
df = pd.read_csv('Restaurant_Reviews.tsv',delimiter="\t",quoting=QUOTE_NONE)
print(df.shape)
df
```

### Cleaning the Data

*   lower case
*   Removing non word characters
*   Removing all stop words
*   Setmming , bringing to root form



```python
stop_words = set(stopwords.words('english'))
stop_words.discard('not')
stop_words.discard('hate')
stop_words.discard('no')


porter_stemmer = PorterStemmer()

corpus= []
for index, row in df.iterrows():
  review=row['Review']
  review = re.sub("^[a-zA-Z]", "", review)
  review = re.sub(r"^\d+", "", review)

  review=review.lower()
  review= " ".join([porter_stemmer.stem(word)  for word in review.split() if word not in stop_words])
  corpus.append(review)
```

```python
print(len(corpus))
corpus[:10]
```

###Implementing Bag of Words model || convert words to numerical vectors

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
x= cv.fit_transform(corpus).toarray()
y=df.iloc[:,-1].values
print(len(cv.get_feature_names_out()))
print(len(x[0]))
```

```python
x
```

```python
print(cv.get_feature_names_out())

print(len(cv.get_feature_names_out()))
```

### split into Training and test set

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state=0,test_size=0.2)
```

```python
x_train
```

```python
y_train
```

### Training Logistic regression Classification Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

classifier = LogisticRegression(random_state=42)
classifier.fit(x_train, y_train)
```

### Predicting Results Test Set

```python
y_pred= classifier.predict(x_test)
```

### Confusion Matrix

```python
confusion_matrix= confusion_matrix(y_test, y_pred)
confusion_matrix
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(y_test, y_pred)
```
