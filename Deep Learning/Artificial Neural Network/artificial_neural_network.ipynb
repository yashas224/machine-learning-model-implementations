## imports

```python
import pandas as pd
import numpy as np
import tensorflow as tf
import keras
from keras import layers

from sklearn.model_selection import train_test_split

```

## Data Preprocessing

### importing  Datasets

```python
df = pd.read_csv('Churn_Modelling.csv')
print(df.shape)
df
```

```python
x= df.iloc[:,3:-1].values
y=df.iloc[:,-1].values
```

```python
x
```

```python
y
```

### Encoding Categorical Data

#### Label encoding of gender column

```python
x[:,2]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
x[:,2]=le.fit_transform(x[:,2])
```

```python
x
```

#### One Hot encoding of Geography column

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
x=np.array(ct.fit_transform(x))
print(ct.get_feature_names_out())
```

```python
x
```

#### Split data into training  and test sets

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=0)
```

```python
x_train
```

#### Feature scaling

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_train =scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
x_train
```

## Building ANN

```python
def displayNeuralNetwork(model):
  print(model.summary())
  print("----------------------------------")
  print(model.get_weights())
```

### initializing ANN

```python
ann = keras.Sequential()
```

### Adding input layer and first Hidden Layer

```python
len(x_train[0])
```

```python
ann.add(keras.Input(shape=(len(x_train[0]),)))
```

```python
ann.add(layers.Dense(units=6,activation="relu", name="deepLayer-1"))
```

```python
displayNeuralNetwork(ann)
```

### Adding second hidden layer

```python
ann.add(layers.Dense(units=6,activation="relu", name="deepLayer-2"))
```

```python
displayNeuralNetwork(ann)
```

### Adding output layer

```python
ann.add(layers.Dense(units=1,activation="sigmoid", name="output_layer"))
```

```python
displayNeuralNetwork(ann)
```

## Training ANN

### Compiline ANN

```python
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

### Training ANN

```python
ann.fit(
    x_train,
    y_train,
    batch_size=32,
    epochs=25
)
```

```python
displayNeuralNetwork(ann)
```
