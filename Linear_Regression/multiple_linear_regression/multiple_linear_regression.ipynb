### libraries

```python
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt
```

### Import Data Set

```python
df = pd.read_csv("50_Startups.csv")
print("Data")
print(df)

x= df.iloc[:,:-1].values
y=df.iloc[:,-1].values

print("Features")

print(x)

print("Dependent Variable")

print(y)
```

###Encoding the Independent Variable

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [3])], remainder='passthrough')
x=np.array(ct.fit_transform(x))
type(x)
```

```python
ct.get_feature_names_out()
```

No need to worry about the dummy varoable trap - the library manages it

```python
x
```

### Splitting the dataset into the Training set and Test set


```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.2,random_state=1)
```

### Training multiple linear regression model on the training set

```python
print("training set features")

print(X_train)
print("training set dependent variables")
print(y_train)
```

```python
from sklearn import linear_model
regressor=linear_model.LinearRegression()
regressor.fit(X_train,y_train)
# fit methd trains the model on the training set
```

###Predicting Test Set resukts

```python
print("Test Set features")
print(X_test)
print("Test Set Dependent variable")
print(y_test)

y_predict=regressor.predict(X_test)
```

```python
y_predict
```

###Comparing Test vs predicted profits

```python
comparison = np.concatenate( ( y_test.reshape(-1,1 ) ,y_predict.reshape(-1,1) ), axis=1)
```
